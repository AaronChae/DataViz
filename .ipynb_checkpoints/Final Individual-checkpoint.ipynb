```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, widgets
```

```python
df = pd.read_csv("/Users/myeong-won/Desktop/Final Project/1. Individual/Existing_Buildings_Energy_Performance_Ordinance_Report.csv")
```

```python
df
```

```python
buildings_2022 = df[df['Benchmark Year'] == 2022]
buildings_2022
```

```python
cleaned_scores = buildings_2022['ENERGY STAR Score'].dropna()
buildings_2022 = buildings_2022.dropna(subset=['ENERGY STAR Score'])
```

```python
buildings_2022
```

```python
buildings_2022.columns
```

```python
selected_columns = ['Building Address', 'Postal Code', 'Category',
                    'Percent Better than National Median Site EUI',
                    'Property Type - Self Selected', 'Building Name', 
                    'Benchmark Year', 'ENERGY STAR Score', 
                    'Site EUI (kBtu/ft2)', 'Source EUI (kBtu/ft2)']

buildings_2022_selected = buildings_2022[selected_columns]
```

```python
buildings_2022_selected
```

```python
buildings_2022_selected.to_csv('/Users/myeong-won/Desktop/Final Project/buildings_2022_selected.csv', index=False)
```

```python
building_name_dropdown = widgets.Dropdown(
    options=buildings_2022_selected['Building Name'].unique(),
    value=buildings_2022_selected['Building Name'].iloc[0],
    description='Building Name:',
    disabled=False,
)

def plot_energy_star_score(building_name):
    selected_building_scores = buildings_2022_selected.loc[
        buildings_2022_selected['Building Name'] == building_name,
        'ENERGY STAR Score'
    ].values

    if len(selected_building_scores) == 0:
        print(f"Building with the name '{building_name}' not found.")
    else:
        selected_building_score = selected_building_scores.mean()
        all_scores = buildings_2022_selected['ENERGY STAR Score']
        percentile = (all_scores < selected_building_score).mean() * 100

        plt.figure(figsize=(10, 6))
        plt.hist(all_scores, bins=30, color='lightblue', edgecolor='black')
        plt.axvline(selected_building_score, color='red', linestyle='dashed', linewidth=2, label=f'{building_name} Score')
        
        plt.annotate(f'{building_name} Score', xy=(selected_building_score, 0), xytext=(selected_building_score, 20),
                     arrowprops=dict(facecolor='red', shrink=0.05),
                     fontsize=10, color='black', ha='center')
        
        plt.title('2022 ENERGY STAR Score Distribution')
        plt.xlabel('ENERGY STAR Score')
        plt.ylabel('Frequency')
        plt.legend()
        plt.show()

        print(f"{building_name}'s ENERGY STAR Score: {selected_building_score}")
        print(f"{building_name}'s Score is higher than {percentile:.2f}% of the buildings.")

interact(plot_energy_star_score, building_name=building_name_dropdown)
```

```python
building_name_dropdown = widgets.Dropdown(
    options=buildings_2022_selected['Building Name'].unique(),
    value=buildings_2022_selected['Building Name'].iloc[0],
    description='Building Name:',
    disabled=False,
)

def plot_energy_star_score(building_name):
    selected_building_scores = buildings_2022_selected.loc[
        buildings_2022_selected['Building Name'] == building_name,
        'ENERGY STAR Score'
    ].values

    if len(selected_building_scores) == 0:
        print(f"Building with the name '{building_name}' not found.")
    else:
        selected_building_score = selected_building_scores.mean()
        all_scores = buildings_2022_selected['ENERGY STAR Score']
        percentile = (all_scores < selected_building_score).mean() * 100
        score_counts = buildings_2022_selected['ENERGY STAR Score'].value_counts().sort_index()
        
        plt.figure(figsize=(10, 6))
        
        plt.plot(score_counts.index, score_counts.values, marker='o', linestyle='-', color='lightblue', markersize=5, label='Other Buildings')        
        
        plt.scatter(selected_building_score, score_counts.loc[selected_building_score], color='red', marker='o', s=50, label=f'{building_name} Score')
        
        plt.annotate(f'{building_name} Score', xy=(selected_building_score, score_counts.loc[selected_building_score]), xytext=(selected_building_score, score_counts.max() - 30),
                     arrowprops=dict(facecolor='red', shrink=0.8),  # shrink 값으로 꼬리의 길이 조절
                     fontsize=10, color='black', ha='center')
        
        annotate_text = f"{building_name}'s Score: {selected_building_score}\nHigher than {percentile:.2f}% of buildings."
        plt.text(selected_building_score, score_counts.max() - 50, annotate_text, ha='center', va='center', color='black', fontsize=10)
        
        plt.title('2022 ENERGY STAR Score Distribution')
        plt.xlabel('ENERGY STAR Score')
        plt.ylabel('Frequency')
        plt.legend()
        plt.show()

interact(plot_energy_star_score, building_name=building_name_dropdown)
```
